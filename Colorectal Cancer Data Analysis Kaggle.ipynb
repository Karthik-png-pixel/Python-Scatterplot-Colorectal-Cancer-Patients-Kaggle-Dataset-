```python
# Loading the library
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Importing the dataset
df = pd.read_csv('Colorectal Cancer Patient Data.csv')
```

```python
# Exploring the Data
print(df)
df.head()
```

```python
# Summary Statistics
df.describe()
```

```python
# Identifying and Removing Null values
df.isnull()

df.dropna()
```

```python
# Relationship between Age and DFS among Colorectal Cancer Patients (Male and Female)

## Separate by Gender
male_df = df[df["Gender"] == "Male"]
female_df = df[df["Gender"] == "Female"]

```

```python
summary = df.groupby("Gender")[["Age (in years)", "DFS (in months)"]].mean()
print(summary)
```

```python
# Visualization Using Scatterplot
plt.figure(figsize = (8,5))
```

```python
plt.scatter(male_df["Age (in years)"], male_df["DFS (in months)"],
                color = 'blue', label = 'Male', alpha = 0.7)
plt.scatter(female_df['Age (in years)'], female_df['DFS (in months)'], 
            color='red', label='Female', alpha=0.7)

plt.title("Relationship between Age and DFS among Colorectal Cancer Patients")
plt.xlabel("Age (in years)")
plt.ylabel("DFS (in months)")
plt.legend()
plt.tight_layout()
plt.show()



# Overall correlation between Age and DFS
correlation = df['Age (in years)'].corr(df['DFS (in months)'])
print(f'Correlation between Age and DFS: {correlation:.2f}')
```
